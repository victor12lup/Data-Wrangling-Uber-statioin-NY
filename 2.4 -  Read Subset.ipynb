## 3 - Reading Data Subset

Big Datasets are everywhere. It might however be challanging to read the whole dataset one time. In this case, you need to read subset of the data. In Geopandas, you can use pre-filterinto to load geospatial data with a geometry, bounding box or rows. 


### Learning Objectives:

1. Read subset of data using Geometry.
2. Read subset of data using Row filter.

```python
import geopandas as gpd
import pandas as pd
```

```python
# Read the Countries data (will use the geometry later)

shp = 'data/countries.shp'
countries = gpd.read_file(shp)
countries.head()
```

```python
countries.crs
```

```python
countries.plot();
```

## Reading subset of data using Geometry

```python
countries[countries["CONTINENT"] == "Africa"]
```

```python
countries[countries["CONTINENT"] == "Africa"].plot();
```

```python
# Reading Acled Dataset by subseting Africa

africa_acled = gpd.read_file(
    "output/acled2019.shp",
    mask = countries[countries["CONTINENT"] == "Africa"]
)
africa_acled.plot(markersize=0.5);
```

## Reading subset of data using Rows

```python
acled_subset = gpd.read_file(
    "output/acled2019.shp",
    rows=200
)
acled_subset.plot();
```

```python
acled_subset = gpd.read_file(
    "output/acled2019.shp",
    rows=slice(200, 500)
)
acled_subset.plot();
```

**Recap**
* `mask` --> While reading the data, you can filter out by geometry using mask parameter.
* `rows` --> You can also use rows to limit the number of rows while reading the data.
